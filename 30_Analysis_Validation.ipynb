```python
all_choices = pd.Series([1, 3, 4, 5, 21, 22])
columns_to_sum = ['foot_prob', 'carp_prob', 'card_prob', 'pt_prob', 'cbike_prob', 'ebike_prob']
mode_reweighting = pd.read_csv('../mode-reweighting.csv')
mode_reweighting
```

```python
#kfcv, mode share for all trips

# mode share in the data
modeshare_recorded = simresults_k12345['choice'].value_counts() / len(simresults_k12345)

# mode share in the simulation results, taking most likely mode for each trip
modeshare_sim_ml = simresults_k12345['choice_sim'].value_counts() / len(simresults_k12345)

# mode share in the simulation results, taking probabilities each trip
modeshare_sim_prob = []

for column in columns_to_sum:
    prob = np.sum(simresults_k12345[column]) / len(simresults_k12345)
    modeshare_sim_prob.append(prob)
    
d = {'mode': ['foot','carp','card','pt','cbike','ebike'], 'recorded': [modeshare_recorded[1],modeshare_recorded[3],modeshare_recorded[4],modeshare_recorded[5],modeshare_recorded[21],modeshare_recorded[22]],
     'sim_mostlikely': [modeshare_sim_ml[1],modeshare_sim_ml[3],modeshare_sim_ml[4],modeshare_sim_ml[5],modeshare_sim_ml[21],modeshare_sim_ml[22]],
    'sim_prob': [modeshare_sim_prob[0],modeshare_sim_prob[1],modeshare_sim_prob[2],modeshare_sim_prob[3],modeshare_sim_prob[4],modeshare_sim_prob[5]]}

overview_unweighted = pd.DataFrame(data=d)

overview_unweighted.to_csv('../validation/modeshares_kfcv_alltrips.csv')
overview_unweighted
```
