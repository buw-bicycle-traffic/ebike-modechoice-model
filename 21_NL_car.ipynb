```python
results = res.bioResults(pickle_file='Estimation results/Final//ModeChoice-MiD_NL_car.pickle')

# point elasticity for change in infra and impact on cbike
point_elas_cbike_infra = Derive(prob_Opt21, 'cyclinfra') * cyclinfra / prob_Opt21

simulate = {
    'weight': weight,
    'cbike_prob': prob_Opt21,
    'Weighted cbike_prob': prob_Opt21 * weight,
    'Point elast. cbike infra': point_elas_cbike_infra,
}

biosim = bio.BIOGEME(database_complete, simulate)
biosim.modelName = 'point elas. cbike infra'
simresults = biosim.simulate(results.get_beta_values())

aggregate_point_elas_cbike_infra = (
    simresults['Weighted cbike_prob']
    * simresults['Point elast. cbike infra']
    / simresults['Weighted cbike_prob'].sum()
    ).sum()
    
aggregate_point_elas_cbike_infra
```
